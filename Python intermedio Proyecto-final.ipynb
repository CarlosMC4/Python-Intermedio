```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files
```

```python
df.rename(index=str, columns={'Ranking' : 'ranking',
                              'Username' : 'username',
                              'Subscribers' : 'subscribers',
                              'Uploads' : 'uploads',
                              'Views' : 'views',
                              'Country' : 'country'}, inplace=True)
```

```python
df.head()
```

```python
df['username'] = df.username.str.lower()
```

```python
df.head()
```

```python
df_new=df
```

```python
df_new = df.dropna()
```

```python
df_new.isnull().sum().sort_values(ascending=False)
```

```python
df_new['subscribers'] = pd.to_numeric(df_new['subscribers'].astype(str).str.replace('M', '').str.replace('K', ''), errors='coerce')
```

```python
df_new.groupby('subscribers')['views'].agg('sum').plot(kind="bar")
```

```python
df_new.to_csv('YOUTUBE_CHANNELS_PROCESADO.csv', index=False)
```
